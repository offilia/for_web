```python
#必要なライブラリをインポートする
from keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from keras.preprocessing import image
import numpy as np
```

```python
#モデルの呼び出しと確認
model = VGG16()
model.summary()
```

```python
#インプット画像を224 x 224にリサイズして画像の読み込み
img_1 = image.load_img('23831869_s.jpg', target_size=(224, 224))#jpgの前の部分はアップロードした画像の名前！
```

```python
#読み込んだ画像をarrayに変換
arr_1 = image.img_to_array(img_1)
#画像をVGG16モデルの事前学習時と同じ状態に合わせて変換
arr_1 = preprocess_input(arr_1)
# 3次元テンソル（rows, cols, channels) を 4次元テンソル (samples, rows, cols, channels) に変換
input = np.stack([arr_1])
input.shape
```

```python
#予測値を算出
#予想値は確率が高いTOP5を表示
predict = model.predict(input)
result = decode_predictions(predict, top=5)
print(result)
```
